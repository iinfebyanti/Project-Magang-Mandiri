```python
import re
# Proses pembersihan teks
dataclean = []
for line in pre_text:
    line = re.sub(r"http\S+", " ", line)            # Menghapus link
    line = re.sub(r"@\S+", " ", line)               # Menghapus username
    line = re.sub(r"#\S+", " ", line)               # Menghapus hashtag
    line = re.sub(r"\d", " ", line)                 # Menghapus angka
    line = re.sub(r"<.*?>", " ", line)              # Menghapus emotikon/HTML tag
    line = re.sub(r"\n", " ", line)                 # Menghapus baris baru
    line = re.sub(r",", " ", line)                  # Menghapus koma
    line = re.sub(r"[^\w\s]", " ", line)            # Menghapus tanda baca
    line = re.sub(r"\s+", " ", line).strip()        # Menghapus spasi berlebih
    dataclean.append(line)
    
#Case folding
data_lower=[]
for line in dataclean:
    a = line.lower()
    data_lower.append(a)

#Normalisiasi
from collections import OrderedDict
def replace(text, dic):
    for typo, correct in dic.items():
        text = re.sub(r'\b{}\b'.format(re.escape(typo)), correct, text)
    return text
dic = OrderedDict(kata)
datachange = []
for line in data_lower:
    result = replace(line, dic)
    datachange.append(result)
    
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
#Stemming
factory=StemmerFactory()
stemmer=factory.create_stemmer()
datastemmed=map(lambda x: stemmer.stem(x), datachange)
datastemmed=list(datastemmed)

#Stopwords
from nltk.tokenize import word_tokenize
with open(r"C:\Users\IIN FEBYANTI\OneDrive\Sains Data\Semester 5\NLP\Stopwords.txt", encoding='utf-8') as f:
    stopwords = f.read().splitlines()
databersih = [
    " ".join([word for word in word_tokenize(kalimat) if word.lower() not in stopwords])
    for kalimat in datastemmed
]

df_stem = pd.DataFrame({
    'Sebelum Cleaning Text': pre_text,
    'Setelah Cleaning Text': databersih,
})

df_stem.head()
```

```python
import pandas as pd

df_label = data[['databersih', 'score']].copy()

# Tambahkan kolom baru 'label' berdasarkan nilai score
df_label['label'] = df_label['score'].apply(lambda x: 'positif' if x >= 4 else 'negatif')

# Tampilkan DataFrame
df_label.head()
```

```python
# ============================================================
# 1. IMPORT LIBRARY
# ============================================================

import matplotlib.pyplot as plt

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import StratifiedKFold
from sklearn.svm import SVC

from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, roc_auc_score, roc_curve,
    confusion_matrix, ConfusionMatrixDisplay
)

# PERSIAPAN DATA
import numpy as np
import pandas as pd
# Encoding label
df_label['label'] = df_label['label'].map({
    'negatif': 0,
    'positif': 1
})

X_text = df_label['databersih']
y = df_label['label']

# STRATIFIED K-FOLD SETUP
from sklearn.model_selection import StratifiedKFold
skf = StratifiedKFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)

metrics = {
    'Accuracy': [],
    'Precision': [],
    'Recall': [],
    'F1': [],
    'ROC AUC': []
}

y_true_all = []
y_pred_all = []

base_fpr = np.linspace(0, 1, 101)
tprs = []

# TRAINING & EVALUASI (STRATIFIED K-FOLD)
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, roc_auc_score, roc_curve,
    confusion_matrix, ConfusionMatrixDisplay
)

for fold, (train_idx, test_idx) in enumerate(skf.split(X_text, y), 1):
    print(f"Fold ke-{fold}")

    X_train_text = X_text.iloc[train_idx]
    X_test_text  = X_text.iloc[test_idx]
    y_train = y.iloc[train_idx]
    y_test  = y.iloc[test_idx]

    # TF-IDF (FIT HANYA DI DATA TRAIN)
    tfidf = TfidfVectorizer()
    X_train = tfidf.fit_transform(X_train_text)
    X_test  = tfidf.transform(X_test_text)

    # SVM LINEAR BALANCED
    model = SVC(
        kernel='linear',
        probability=True,
        class_weight='balanced',
        random_state=42
    )

    model.fit(X_train, y_train)

    # Prediksi
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1]

    y_true_all.extend(y_test)
    y_pred_all.extend(y_pred)

    # Hitung metrik
    metrics['Accuracy'].append(accuracy_score(y_test, y_pred))
    metrics['Precision'].append(precision_score(y_test, y_pred, zero_division=0))
    metrics['Recall'].append(recall_score(y_test, y_pred, zero_division=0))
    metrics['F1'].append(f1_score(y_test, y_pred, zero_division=0))
    metrics['ROC AUC'].append(roc_auc_score(y_test, y_proba))

    # ROC Curve per fold
    fpr, tpr, _ = roc_curve(y_test, y_proba)
    tpr_interp = np.interp(base_fpr, fpr, tpr)
    tpr_interp[0] = 0.0
    tprs.append(tpr_interp)

# TABEL HASIL EVALUASI (RATA-RATA K-FOLD)
metrics_df = pd.DataFrame({
    'Model': ['SVM Linear (5-Fold CV)'],
    'Accuracy': [np.mean(metrics['Accuracy'])],
    'Precision': [np.mean(metrics['Precision'])],
    'Recall': [np.mean(metrics['Recall'])],
    'F1-Score': [np.mean(metrics['F1'])],
    'ROC AUC': [np.mean(metrics['ROC AUC'])]
}).round(4)

print("\nHasil Evaluasi Rata-rata 5-Fold:")
print(metrics_df)

# ROC CURVE (MEAN ROC K-FOLD)
import matplotlib.pyplot as plt

mean_tpr = np.mean(tprs, axis=0)
mean_tpr[-1] = 1.0
mean_auc = np.mean(metrics['ROC AUC'])

plt.figure(figsize=(7,6))
plt.plot(base_fpr, mean_tpr, label=f'Mean ROC (AUC = {mean_auc:.3f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve - SVM Linear')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()

# CONFUSION MATRIX (HASIL K-FOLD)
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm_cv = confusion_matrix(y_true_all, y_pred_all)

disp_cv = ConfusionMatrixDisplay(
    confusion_matrix=cm_cv,
    display_labels=['Negatif', 'Positif']
)

disp_cv.plot(cmap='Blues', values_format='d')
plt.title('Confusion Matrix - Stratified 5-Fold CV')
plt.show()

# MODEL FINAL (TRAIN SELURUH DATA)
tfidf_final = TfidfVectorizer()
X_all = tfidf_final.fit_transform(X_text)

final_model = SVC(
    kernel='linear',
    probability=True,
    class_weight='balanced',
    random_state=42
)

final_model.fit(X_all, y)

y_all_pred = final_model.predict(X_all)

cm_all = confusion_matrix(y, y_all_pred)

disp_all = ConfusionMatrixDisplay(
    confusion_matrix=cm_all,
    display_labels=['Negatif', 'Positif']
)

disp_all.plot(cmap='Blues', values_format='d')
plt.title('Confusion Matrix - Final Model')
plt.show()
```
