```python
# Siapkan Data
import pandas as pd
import numpy as np
df_hasil = data1[['Kab/Kota', 'Hasil Balai',
                 'Petugas_1', 'Petugas_2', 'Petugas_3',
                 'Petugas_4', 'Petugas_5']].copy()

df_hasil['Hasil Balai'] = df_hasil['Hasil Balai'].fillna('Tidak Ada Data')
print (df_hasil)

# Ubah ke Format Long
df_long = pd.melt(
    df_hasil,
    id_vars=['Kab/Kota', 'Hasil Balai'],
    value_vars=['Petugas_1', 'Petugas_2', 'Petugas_3', 'Petugas_4', 'Petugas_5'],
    var_name='KolomPetugas',
    value_name='Petugas'
)

df_long = df_long.dropna(subset=['Petugas', 'Kab/Kota'])
df_long['Petugas'] = df_long['Petugas'].astype(str).str.strip()
df_long['Kab/Kota'] = df_long['Kab/Kota'].astype(str).str.strip()
df_long['Hasil Balai'] = df_long['Hasil Balai'].astype(str).str.strip()
print(df_long)

# Hitung Frekuensi per Wilayah dan Hasil
df_long['Wilayah_Hasil'] = df_long['Kab/Kota'] + " | " + df_long['Hasil Balai']
df_freq = df_long.groupby(['Petugas', 'Wilayah_Hasil']).size().reset_index(name='Jumlah')
print(df_freq)

# Pivot: Petugas x Wilayah_Hasil
df_pivot = df_freq.pivot_table(index='Petugas', columns='Wilayah_Hasil', values='Jumlah', fill_value=0)
print(df_pivot)

# Normalisasi
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_pivot)

# PCA 
from sklearn.decomposition import PCA
pca = PCA(n_components=None)
X_pca = pca.fit_transform(X_scaled)

eigenvalues = pca.explained_variance_
loadings = pd.DataFrame(
    pca.components_,
    columns=df_pivot.columns,
    index=[f'PC{i+1}' for i in range(len(eigenvalues))]
)

# Evaluasi K-Means dengan Elbow & Silhouette
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
inertia, silhouette_scores = [], []
K = range(2, 8)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X_pca)
    inertia.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X_pca, labels))

# Visualisasi Elbow & Silhouette
import matplotlib.pyplot as plt
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].plot(K, inertia, marker='o')
axes[0].set_title('Elbow Method (Inertia) - PCA Based')
axes[0].set_xlabel('Jumlah Cluster')
axes[0].set_ylabel('Inertia')
axes[0].grid(alpha=0.3)

axes[1].plot(K, silhouette_scores, marker='o', color='green')
axes[1].set_title('Silhouette Score - PCA Based')
axes[1].set_xlabel('Jumlah Cluster')
axes[1].set_ylabel('Silhouette Score')
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.show()

# K-Means Final Menggunakan Semua PCA
n_cluster = 2
kmeans = KMeans(n_clusters=n_cluster, random_state=42, n_init=10)
df_pivot['Cluster'] = kmeans.fit_predict(X_pca)

sil_score = silhouette_score(X_pca, df_pivot['Cluster'])
inertia_final = kmeans.inertia_

print(f"\n Evaluasi Model Final (n={n_cluster}):")
print("----------------------------------------")
print(f"• Inertia            : {inertia_final:.4f}")
print(f"• Silhouette Score   : {sil_score:.4f}")

print("\n📌 Daftar Petugas per Cluster:")
print(df_pivot[['Cluster']].sort_values('Cluster'))

# PCA 2D untuk Visualisasi
import seaborn as sns
pca_2d = PCA(n_components=2)
X_pca_2d = pca_2d.fit_transform(X_scaled)

df_vis = pd.DataFrame(X_pca_2d, columns=['PC1', 'PC2'])
df_vis['Cluster'] = df_pivot['Cluster'].values.astype(str)
df_vis['Petugas'] = df_pivot.index

plt.figure(figsize=(10,7))
sns.scatterplot(
    x='PC1', y='PC2', hue='Cluster',
    data=df_vis, palette='Set2', s=100, edgecolor='black'
)

for i, txt in enumerate(df_vis['Petugas']):
    plt.annotate(txt, (df_vis.PC1[i]+0.05, df_vis.PC2[i]+0.05), fontsize=8)

plt.title(f'PCA 2D - Distribusi Petugas Berdasarkan Cluster (n={n_cluster})')
plt.xlabel('Komponen Utama 1')
plt.ylabel('Komponen Utama 2')
plt.legend(title='Cluster')
plt.tight_layout()
plt.show()

# Pie Chart Persentase Petugas per Cluster
cluster_count = df_pivot['Cluster'].value_counts().sort_index()

plt.figure(figsize=(6,6))
plt.pie(
    cluster_count.values, 
    labels=[f'Cluster {i}' for i in cluster_count.index], 
    autopct='%1.1f%%',
    colors=plt.cm.Pastel2(np.linspace(0, 1, len(cluster_count))),
    startangle=90
)
plt.title('Persentase Petugas per Cluster')
plt.axis('equal')
plt.show()

# Proporsi Hasil Balai per Cluster
df_merge = df_long.merge(df_pivot[['Cluster']], left_on='Petugas', right_index=True, how='inner')

count_hasil = (
    df_merge.groupby(['Cluster', 'Hasil Balai'])
    .size()
    .reset_index(name='Count')
)
count_hasil['Persentase'] = count_hasil.groupby('Cluster')['Count'].transform(lambda x: 100 * x / x.sum())

count_hasil_nonzero = count_hasil[count_hasil['Persentase'] > 0]

plt.figure(figsize=(10,6))
ax = sns.barplot(
    x='Cluster', 
    y='Persentase', 
    hue='Hasil Balai', 
    data=count_hasil_nonzero, 
    palette='Set3'
)

for p in ax.patches:
    height = p.get_height()
    if np.isfinite(height) and height > 0:
        ax.text(
            p.get_x() + p.get_width()/2,
            height + 0.5,
            f"{height:.1f}%",
            ha='center', fontsize=9, weight='bold'
        )

plt.title('Proporsi Hasil Balai per Cluster')
plt.tight_layout()
plt.show()

# Visualisasi Wilayah Pemeriksaan per Cluster
wilayah_summary = (
    df_merge.groupby(['Cluster', 'Kab/Kota'])
    .size()
    .reset_index(name='Frekuensi')
)
wilayah_summary['Persentase'] = wilayah_summary.groupby('Cluster')['Frekuensi'].transform(lambda x: 100 * x / x.sum())
wilayah_summary = wilayah_summary.replace([np.inf, -np.inf], np.nan).dropna(subset=['Persentase'])

total_persen_per_kota = wilayah_summary.groupby('Kab/Kota')['Persentase'].sum().sort_values(ascending=False)
wilayah_summary['Kab/Kota'] = pd.Categorical(wilayah_summary['Kab/Kota'], categories=total_persen_per_kota.index, ordered=True)

plt.figure(figsize=(16,11))
ax = sns.barplot(
    x='Persentase',
    y='Kab/Kota',
    hue='Cluster',
    data=wilayah_summary,
    dodge=True,
    palette='coolwarm'
)

for p in ax.patches:
    width = p.get_width()
    if np.isfinite(width) and width > 0:
        plt.text(
            width + 0.3,
            p.get_y() + p.get_height() / 2,
            f"{width:.1f}%",
            ha='left', va='center', fontsize=12, weight='bold'
        )

plt.title('Persentase Pemeriksaan per Wilayah dan Cluster')
plt.tight_layout()
plt.show()
```
